1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.


```python
class matrix:

    def __init__(self, *args):
        
        if len(args) == 2:   ##zero matrix
            n=args[0]
            m=args[1]
            
            self.rows=n
            self.cols=m
            
            self.data = []
            for i in range(n):
                row = []
                for j in range(m):
                    row.append(0)
                self.data.append(row)
        
        
        elif len(args)==1:  ##list of lists
            values = args[0]
            
            row_length=len(values[0])
            for row in values:
                if len(row)!=row_length:
                    raise ValueError("Matrix rows must have same length")
            
            self.rows=len(values)
            self.cols=row_length
            self.data=values
        
        else:
            raise ValueError("Invalid initialization")
    
    
    def __getitem__(self, index):
        
        if isinstance(index, tuple):  #M[i,j]
            i, j = index
            return self.data[i][j]
        
        return self.data[index]   #M[i]
    
    
    def __setitem__(self, index, value):
        
        if isinstance(index, tuple):
            i,j= index
            self.data[i][j] = value
        else:
            self.data[index] = value
    
    
    def assign(self, other): 
        
        if isinstance(other, matrix):
            
            if self.rows != other.rows or self.cols != other.cols:
                raise ValueError("Matrix sizes must match")
            
            for i in range(self.rows):
                for j in range(self.cols):
                    self.data[i][j] = other.data[i][j]
        
        
        elif isinstance(other, list):
            
            if len(other) != self.rows or len(other[0]) != self.cols:
                raise ValueError("Matrix sizes must match")
            
            for i in range(self.rows):
                for j in range(self.cols):
                    self.data[i][j] = other[i][j]
        
        else:
            raise ValueError("Invalid assignment type")
```

```python
M1 = matrix(2,3)
print(M1.data)

M2 = matrix([[1,2,3],[4,5,6]])
print(M2[0][1])
print(M2[1,2])

M3 = matrix(2,3)
M3.assign(M2)
print(M3.data)

M3.assign([[7,8,9],[10,11,12]])
print(M3.data)
```

2. Add the following methods:
    * `shape()`: returns a tuple `(n,m)` of the shape of the matrix.
    * `transpose()`: returns a new matrix instance which is the transpose of the matrix.
    * `row(n)` and `column(n)`: that return the nth row or column of the matrix M as a new appropriately shaped matrix object.
    * `to_list()`: which returns the matrix as a list of lists.
    *  `block(n_0,n_1,m_0,m_1)` that returns a smaller matrix located at the n_0 to n_1 columns and m_0 to m_1 rows. 
    * Modify `__getitem__` implemented above to support slicing.
        

```python
def shape(self):
    return (self.rows, self.cols)


def transpose(self):
    result=[]
    for j in range(self.cols):
        row=[]
        for i in range(self.rows):
            row.append(self.data[i][j])
        result.append(row)
    return matrix(result)


def row(self,n):
    return matrix([self.data[n]])


def column(self,n):
    col=[]
    for i in range(self.rows):
        col.append([self.data[i][n]])
    return matrix(col)


def to_list(self):
    return self.data


def block(self,n_0,n_1,m_0,m_1):
    result=[]
    for i in range(n_0,n_1):
        row=[]
        for j in range(m_0, m_1):
            row.append(self.data[i][j])
        result.append(row)
    return matrix(result)
```

```python
M = matrix([[1,2,3],
            [4,5,6],
            [7,8,9]])

print(M.shape())
print(M.transpose().data)
print(M.row(1).data)
print(M.column(2).data)
print(M.to_list())
print(M.block(0,2,0,2).data)
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n, m, c):
    data = []
    for i in range(n):
        row = []
        for j in range(m):
            row.append(float(c))
        data.append(row)
    return matrix(data)


def zeros(n, m):
    return constant(n, m, 0.0)


def ones(n, m):
    return constant(n, m, 1.0)


def eye(n):
    data = []
    for i in range(n):
        row = []
        for j in range(n):
            if i == j:
                row.append(1.0)
            else:
                row.append(0.0)
        data.append(row)
    return matrix(data)
```

```python
print(constant(2,3,5).data)
print(zeros(2,2).data)
print(ones(2,2).data)
print(eye(3).data)
```

4. Add the following member functions to your class. Make sure to appropriately test the dimensions of the matrices to make sure the operations are correct.
    * `M.scalarmul(c)`: a matrix that is scalar product $cM$, where every element of $M$ is multiplied by $c$.
    * `M.add(N)`: adds two matrices $M$ and $N$. Don’t forget to test that the sizes of the matrices are compatible for this and all other operations.
    * `M.sub(N)`: subtracts two matrices $M$ and $N$.
    * `M.mat_mult(N)`: returns a matrix that is the matrix product of two matrices $M$ and $N$.
    * `M.element_mult(N)`: returns a matrix that is the element-wise product of two matrices $M$ and $N$.
    * `M.equals(N)`: returns true/false if $M==N$.

```python
class matrix:

    def __init__(self, values):
        self.data = values
        self.rows = len(values)
        self.cols = len(values[0])

    def scalarmul(self, c):
        result = []
        for i in range(self.rows):
            row = []
            for j in range(self.cols):
                row.append(self.data[i][j] * c)
            result.append(row)
        return matrix(result)

    def add(self, N):
        if self.rows != N.rows or self.cols != N.cols:
            raise ValueError("Sizes must match")
        result = []
        for i in range(self.rows):
            row = []
            for j in range(self.cols):
                row.append(self.data[i][j] + N.data[i][j])
            result.append(row)
        return matrix(result)

    def sub(self, N):
        if self.rows != N.rows or self.cols != N.cols:
            raise ValueError("Sizes must match")
        result = []
        for i in range(self.rows):
            row = []
            for j in range(self.cols):
                row.append(self.data[i][j] - N.data[i][j])
            result.append(row)
        return matrix(result)

    def element_mult(self, N):
        if self.rows != N.rows or self.cols != N.cols:
            raise ValueError("Sizes must match")
        result = []
        for i in range(self.rows):
            row = []
            for j in range(self.cols):
                row.append(self.data[i][j] * N.data[i][j])
            result.append(row)
        return matrix(result)

    def mat_mult(self, N):
        if self.cols != N.rows:
            raise ValueError("Inner dimensions must match")
        result = []
        for i in range(self.rows):
            row = []
            for j in range(N.cols):
                s = 0
                for k in range(self.cols):
                    s = s + self.data[i][k] * N.data[k][j]
                row.append(s)
            result.append(row)
        return matrix(result)

    def equals(self, N):
        if self.rows != N.rows or self.cols != N.cols:
            return False
        for i in range(self.rows):
            for j in range(self.cols):
                if self.data[i][j] != N.data[i][j]:
                    return False
        return True
```

5. Overload python operators to appropriately use your functions in 4 and allow expressions like:

```python
#Used LLM to make a short outline to follow for this question
class matrix:

    def __init__(self, values):
        self.data = values
        self.rows = len(values)
        self.cols = len(values[0])

    def scalarmul(self, c):
        result = []
        for i in range(self.rows):
            row = []
            for j in range(self.cols):
                row.append(self.data[i][j] * c)
            result.append(row)
        return matrix(result)

    def add(self, N):
        if self.rows != N.rows or self.cols != N.cols:
            raise ValueError("Sizes must match")
        result = []
        for i in range(self.rows):
            row = []
            for j in range(self.cols):
                row.append(self.data[i][j] + N.data[i][j])
            result.append(row)
        return matrix(result)

    def sub(self, N):
        if self.rows != N.rows or self.cols != N.cols:
            raise ValueError("Sizes must match")
        result = []
        for i in range(self.rows):
            row = []
            for j in range(self.cols):
                row.append(self.data[i][j] - N.data[i][j])
            result.append(row)
        return matrix(result)

    def element_mult(self, N):
        if self.rows != N.rows or self.cols != N.cols:
            raise ValueError("Sizes must match")
        result = []
        for i in range(self.rows):
            row = []
            for j in range(self.cols):
                row.append(self.data[i][j] * N.data[i][j])
            result.append(row)
        return matrix(result)

    def mat_mult(self, N):
        if self.cols != N.rows:
            raise ValueError("Inner dimensions must match")
        result = []
        for i in range(self.rows):
            row = []
            for j in range(N.cols):
                s = 0
                for k in range(self.cols):
                    s = s + self.data[i][k] * N.data[k][j]
                row.append(s)
            result.append(row)
        return matrix(result)

    def equals(self, N):
        if self.rows != N.rows or self.cols != N.cols:
            return False
        for i in range(self.rows):
            for j in range(self.cols):
                if self.data[i][j] != N.data[i][j]:
                    return False
        return True

    def __add__(self, N):
        return self.add(N)

    def __sub__(self, N):
        return self.sub(N)

    def __mul__(self, other):
        if isinstance(other, (int, float)):
            return self.scalarmul(other)
        return self.mat_mult(other)

    def __rmul__(self, other):
        if isinstance(other, (int, float)):
            return self.scalarmul(other)

    def __eq__(self, N):
        return self.equals(N)
```
