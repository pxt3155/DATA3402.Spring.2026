1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class counter:
    def __init__(self, max_val):
        self.max_val = max_val
        self.value = 0  ##counter = 0
    
    def increment(self):
        if self.value < self.max_val:
            self.value = self.value + 1
        else:
            print("Error: counter is at max value")
            
    def reset(self):
        self.value = 0
        
    def return_value(self):
        return self.value
```

```python
x = counter(3)
print("Initial Value: ", x.return_value())

x.increment()
print("After 1 incriment: ", x.return_value())

x.increment()
x.increment()
x.increment()

x.reset()
print("after reseting: ", x.return_value())
```

```python
class rectangle:
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
        
    def length(self):
        return self.__length
    
    def width(self):
        return self.__width
    
    def x(self):
        return self.__x
    
    def y(self):
        return self.__y
    
    def area(self):
        return (self.__length*self.__width)
    
    def perimeter(self):
        return 2*(self.__length + self.__width)
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
#asked LLM for a short ouline
#outline:
#1. Add contains(x,y) in base shape → prints "Not implemented" and returns False
#2. Rectangle contains → check if point is between left/right and bottom/top edges
#3. Circle contains → check squared distance from center ≤ radius²
#4. Triangle contains → for your right-triangle, use bounding box + line test
#5. Test each class with inside + outside points

class shape:
    
    def area(self):
        print("Not implemented")
    
    def perimeter(self):
        print("Not implemented")
    
    def points(self):
        print("Not implemented")
        return []
    
    def contains(self, x, y):
        print("Not implemented")
        return False


class rectangle:
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
        
    def length(self):
        return self.__length
    
    def width(self):
        return self.__width
    
    def x(self):
        return self.__x
    
    def y(self):
        return self.__y
    
    def area(self):
        return self.__length*self.__width
    
    def perimeter(self):
        return 2*(self.__length+self.__width)
    
    def points(self):
        points = []
        
        x1 = self.__x
        y1 = self.__y
    
        x2 = self.__x+self.__length
        y2 = self.__y
    
        x3 = self.__x+self.__length
        y3 = self.__y+self.__width
    
        x4 = self.__x
        y4 = self.__y+self.__width
    
        points.append((x1, y1))
        points.append((x2, y2))
        points.append((x3, y3))
        points.append((x4, y4))
    
        return points
    
    def contains(self, x, y):
        if (x >= self.__x) and (x <= self.__x+self.__length) and (y >= self.__y) and (y <= self.__y+self.__width):
            ## to check if point (x,y) is inside or on the boundary of the rectangle
            return True
        else:
            return False


class circle:
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y    
        
    def radius(self):
        return self.__radius
    
    def x(self):
        return self.__x
    
    def y(self):
        return self.__y
    
    def area(self):
        return 3.14*(self.__radius*self.__radius)
    
    def perimeter(self):
        return (2*3.14*self.__radius)
    
    def points(self):
        points = []
    
        r = self.__radius
        x = self.__x
        y = self.__y
    
        points.append((x+r, y))  #right
        points.append((x-r, y))  #left
        points.append((x, y+r))  #top
        points.append((x, y-r))  #bottom
    
        return points
    
    def contains(self, x, y):
        dx = x - self.__x
        dy = y - self.__y
        if (dx * dx + dy * dy) <= (self.__radius * self.__radius):
            return True
        else:
            return False


class triangle:
    def __init__(self, a, b, c, base, height, x, y):
        self.__a = a
        self.__b = b
        self.__c = c
        self.__base = base
        self.__height = height
        self.__x = x
        self.__y = y
    
    def a(self):
        return self.__a
    
    def b(self):
        return self.__b
    
    def c(self):
        return self.__c
    
    def base(self):
        return self.__base
    
    def height(self):
        return self.__height
    
    def x(self):
        return self.__x
    
    def y(self):
        return self.__y
    
    def area(self):
        return (0.5*self.__base*self.__height)
    
    def perimeter(self):
        return (self.__a+self.__b+self.__c)
    
    def points(self):
        points = []
    
        x1 = self.__x
        y1 = self.__y
    
        x2 = self.__x+self.__base
        y2 = self.__y
    
        x3 = self.__x
        y3 = self.__y+self.__height
    
        points.append((x1, y1))
        points.append((x2, y2))
        points.append((x3, y3))
    
        return points
    
    def contains(self, x, y):
        x0 = self.__x
        y0 = self.__y
        
        #outside bounding box
        if x < x0 or x > x0+self.__base:
            return False
        if y < y0 or y > y0+self.__height:
            return False
        
        #inside if under the diagonal
        if ((x - x0)/self.__base) +((y - y0)/self.__height) <= 1:
            return True
        else:
            return False
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
#outline:
#1. add overlaps(other) to the base class, print "not implemented"
#2. rectangle overlap for checking if bounding boxes intersect
#3. circle overlap for checking if the circles overlap (if distance bw centers <= sum of radii)
#4. triangle overlap for checking if bounding boxes overlap


class shape:
    
    def area(self):
        print("Not implemented")
    
    def perimeter(self):
        print("Not implemented")
    
    def points(self):
        print("Not implemented")
        return []
    
    def contains(self, x, y):
        print("Not implemented")
        return False
    
    def overlaps(self, other):
        print("Not implemented")
        return False


class rectangle(shape):
    def __init__(self, length, width, x, y):
        self.__length = length
        self.__width = width
        self.__x = x
        self.__y = y
    
    def overlaps(self, other):
        if isinstance(other, rectangle):
            
            if (self.__x+self.__length < other.__x or
                other.__x+other.__length < self.__x or
                self.__y+self.__width < other.__y or
                other.__y+other.__width < self.__y):
                return False
            else:
                return True
        else:
            return False


class circle(shape):
    def __init__(self, radius, x, y):
        self.__radius = radius
        self.__x = x
        self.__y = y
    
    def overlaps(self, other):
        if isinstance(other, circle):
            
            dx = self.__x-other.__x
            dy = self.__y-other.__y
            
            distance_squared = dx*dx + dy*dy
            radius_sum = self.__radius+other.__radius
            
            if distance_squared <= radius_sum*radius_sum:
                return True
            else:
                return False
        else:
            return False


class triangle(shape):
    def __init__(self, base, height, x, y):
        self.__base = base
        self.__height = height
        self.__x = x
        self.__y = y
    
    def overlaps(self, other):
        if isinstance(other, triangle):
            
            if (self.__x+self.__base < other.__x or
                other.__x+other.__base < self.__x or
                self.__y+self.__height < other.__y or
                other.__y+other.__height < self.__y):
                return False
            else:
                return True
        else:
            return False
```

12. Implement the ability to load/save raster drawings and demonstate that your method works. One way to implement this ability:

   * Overload `__repr__` functions of all objects to return strings of the python code that would construct the object.
   
   * In the save method of raster drawing class, store the representations into the file.
   * Write a loader function that reads the file and uses `eval` to instantiate the object.

For example:

```python
class foo:
    def __init__(self,a,b=None):
        self.a=a
        self.b=b
        
    def __repr__(self):
        return "foo("+repr(self.a)+","+repr(self.b)+")"
    
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()
        
   
def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
```
